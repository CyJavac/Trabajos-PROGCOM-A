```python
import random


opcs=["PIEDRA","PAPEL","TIJERAS","LAGARTO","SPOCK"]
      

while True:
    print("- - Opciones - -")
    print("     Piedra")
    print("     Papel")
    print("    Tijeras")
    print("    Lagarto")
    print("     Spock")
    print("")

    ran=random.randrange(0,5)
    cpu= ""
    opc=input("Elige una opción: ")
    print("")
    if opc.upper() == opcs[0]:
        user= "Piedra"
        print("Escogiste:",user)
    elif opc.upper() == opcs[1]:
        user= "Papel"
        print("Escogiste:",user)
    elif opc.upper() == opcs[2]:
        user= "Tijeras"
        print("Escogiste:",user)
    elif opc.upper() == opcs[3]:
        user= "Lagarto"
        print("Escogiste:",user)
    elif opc.upper() == opcs[4]:
        user= "Spock"
        print("Escogiste:",user)
    else:
        print("Error: escoge una opción valida")
        continue
    if ran==0:
        cpu= "Piedra"
    elif ran==1:
        cpu= "Papel"
    elif ran==2:
        cpu= "Tijeras"
    elif ran==3:
        cpu= "Lagarto"
    elif ran==4:
        cpu= "Spock"

    print("Elección CPU:",cpu)

    #Piedra - Papel
    if user=="Piedra" and cpu=="Papel":
        print("Perdiste")
    elif user=="Papel" and cpu=="Piedra":
        print("Ganaste")

    #Piedra - Tijeras
    if user=="Piedra" and cpu=="Tijeras":
        print("Ganaste")
    elif user=="Tijeras" and cpu=="Piedra":
        print("Perdiste")
    
    #Piedra - Spock
    if user=="Piedra" and cpu=="Spock":
        print("Perdiste")
    elif user=="Spock" and cpu=="Piedra":
        print("Ganaste")

    #Piedra - Lagarto
    if user=="Piedra" and cpu=="Lagarto":
        print("Ganaste")
    elif user=="Lagarto" and cpu=="Piedra":
        print("Perdiste")

    #Papel - Tijeras
    if user=="Papel" and cpu=="Tijeras":
        print("Perdiste")
    elif user=="Tijeras" and cpu=="Papel":
        print("Ganaste")

    #Papel - Spock
    if user=="Papel" and cpu=="Spock":
        print("Ganaste")
    elif user=="Spock" and cpu=="Papel":
        print("Perdiste")

    #Papel - Lagarto
    if user=="Papel" and cpu=="Lagarto":
        print("Perdiste")
    elif user=="Lagarto" and cpu=="Papel":
        print("Ganaste")

    #Tijeras - Spock
    if user=="Spock" and cpu=="Tijeras":
        print("Ganaste")
    elif user=="Tijeras" and cpu=="Spock":
        print("Perdiste")

    #Tijeras - Lagarto
    if user=="Lagarto" and cpu=="Tijeras":
        print("Perdiste")
    elif user=="Tijeras" and cpu=="Lagarto":
        print("Ganaste")

    #Spock - Lagarto
    if user=="Spock" and cpu=="Lagarto":
        print("Perdiste")
    elif user=="Lagarto" and cpu=="Spock":
        print("Ganaste")

    if user==cpu:
        print("Empate")

    print("")      
    r=input("Reintentar? Si/No: ")
    if r.upper()=="NO":
        break
    elif r.upper()=="SI":
        continue
    else:
        print("Error: No se escogio una opción valida")
        print("Cerrando...")
        break
```
